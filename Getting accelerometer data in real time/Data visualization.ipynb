# Data visualization

### Importing necessary libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
```

### Reading and exploring the data file

```python
data = pd.read_csv("collision.txt", sep = ',')
data.head()
```

```python
data.describe()
```

```python
print("The data set contains {} rows and {} columns/features".format(data.shape[0], data.shape[1]))
print("The range of the data along {} axis is {}m/s2, along {} axis is {} m/s2 and along {} axis is {} m/s2".format("x", np.abs(data['accel_x'].min() - data['accel_x'].max()),
                                                                                                    "y", np.abs(data[' accel_y'].min() - data[' accel_y'].max()),
                                                                                                     "z", np.abs(data[' accel_z'].min() - data[' accel_z'].max())))
```

### The change of acceleration over time

The acceleration occured along the z-axis (although the range of acceleration values are the least along that axis interestingly), let's observe the change of acceleration along the z-axis and compare it to the change of acceleration along the x-axis.

```python
figure = plt.figure(figsize = (15, 10), facecolor = 'c', edgecolor = 'b', linewidth = 2)
ax = figure.add_axes([0,0,0.75, 0.75])

graph_z = ax.plot(range(468), data[' accel_z'])
graph_x = ax.plot(range(468), data['accel_x'], 'g-')
ax.set_title('Acceleration Against Time')
ax.set_xlabel('t (10ms)')
ax.set_ylabel('acceleration (m/s2)')
ax.legend(labels = ('z-axis', 'x-axis'), loc = 'upper right')
plt.show()
```

Let's see the same graph is separated coordinate systems.

```python
figure = plt.figure(figsize = (12, 12), edgecolor = 'b', linewidth = 2)
ax1 = figure.add_axes([0,0,0.75, 0.75])
ax2 = figure.add_axes([0.6, 0.6, 0.3, 0.3])

graph_z = ax1.plot(range(468), data[' accel_z'])
graph_x = ax2.plot(range(468), data['accel_x'], 'g-')
ax1.set_title('Acceleration Against Time')
ax1.set_xlabel('t (10ms)')
ax1.set_ylabel('acceleration_z (m/s2)')
ax2.set_xlabel('t (10ms)')
ax2.set_ylabel('acceleration_x (m/s2)')
plt.show()
```

### Deeper investigation of the data in 3 separated graphs and focusing on the valuable part of the graphs

```python
# setting figure object
figure = plt.figure(figsize = (14, 12), edgecolor = 'r', linewidth = 5)

# setting axis objects in a grid (2 rows, 3 columns, figures can span across more rows or columns)
ax1 = plt.subplot2grid((2,3), (0,0), rowspan = 2, colspan = 2)
ax2 = plt.subplot2grid((2,3), (0,2), rowspan = 1, colspan = 1)
ax3 = plt.subplot2grid((2,3), (1,2), rowspan = 1, colspan = 1)

#plotting graphs
ax1.plot(range(468), data[' accel_z'])
ax2.plot(range(468), data['accel_x'])
ax3.plot(range(468), data[' accel_y'])

#setting axis labels
ax1.set_xlabel('t (10ms)')
ax2.set_xlabel('t (10ms)')
ax3.set_xlabel('t (10ms)')
ax1.set_ylabel('accel_z (m/s2)')
ax2.set_ylabel(' accel_x (m/s2)')
ax3.set_ylabel('accel_y (m/s2)')

#setting up grids (on graphs)
ax1.grid(color = 'b', ls = '-.', lw = 0.25)
ax2.grid(color = 'y', ls = '-', lw = 0.25)
ax3.grid(color = 'g', ls = '-.', lw = 0.25)

# setting up limits
ax1.set_xlim(75,200)
ax2.set_xlim(75,200)
ax3.set_xlim(75,200)
plt.show()
```
